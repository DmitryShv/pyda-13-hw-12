Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
data_movie = pd.read_csv('ratings.csv')
data_movie.head(5)
```

```python
def rating_class(class_):
    if class_ <= 2:
        return 'low rating'
    elif 2 < class_ <= 4:
        return 'good'
    elif class_ > 4:
        return 'excelent'        
```

```python
# использую название столбца class_ с нижним подчеркиванием вместо class. Так как это совпадает с какой-то внутренней командой
data_movie['class_'] = data_movie['rating'].apply(rating_class)
data_movie.head(20)
```

```python
good_class = data_movie[data_movie.class_ == 'good']
#data_movie.class_.unique()
```
